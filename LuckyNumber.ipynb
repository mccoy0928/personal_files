## 過濾含有數字"4"

```python
def numberFilter(numList):
  tempStr = ""
  for i in numList:
    if i != "4":
      tempStr += i
    else :
      tempStr = "" 
      break 
  return tempStr
```

## 計算數字總和

```python
def numSum(numList):
  tempNum = 0
  for i in numList:
    if i != "4":
      tempNum += int(i)
    else :      
      break 
  return tempNum
```
