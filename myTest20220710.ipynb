## requests & BeautifulSoup 網路爬蟲

```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.iana.org/domains/'
```

```python
myData = requests.get(url)
soup = BeautifulSoup(myData.text)
```

* image.png![image.png]()

```python
#print(myData.text)
print(soup.select('#logo'))
print('\n---------------------\n')
print(soup.find_all('div', id='logo'))
print('\n---------------------\n')
divs = soup.find_all('div')
print(divs[1])
print('\n---------------------\n')
```

```python
len(divs)
```

```python
print(divs[7].find_parent().find_all('li'))  # 讀取find_parent() 父節點
print('\n----------\n')            # 讀取finf_all("li") 標籤資料
print(divs[1].find_parent().find_all('li'))  # 讀取find_parent() 父節點
print('\n----------\n')
```

```python
print(divs[1].find_parent().find_all('li'))  # 讀取find_parent() 父節點
print('\n----------\n')
```

```python
# 從搜尋到的項目裡，尋找父節點裡所有 li 的第三個項目，找到他後方同層的所有 li
print(divs[1].find_parent().find_all('li')[2].find_next_siblings())
print('\n----------\n')
```

```python
# 從搜尋到的項目裡，尋找父節點裡所有 li 的第三個項目，找到他前方同層的所有 li
print(divs[1].find_parent().find_all('li')[2].find_previous_sibling())
```

```python
print(soup('div')[2]) # 比較2種寫法 soup是全部指定 "div"第2個
print('\n----------\n')
print(divs[2])    # 比較2種寫法 divs是先從soup分類後 選定"div"第2個
```

```python
print(soup.find_all('a'))                     # 找出所有 a tag
print(soup.find_all('a', string='Domains'))   # 找出內容字串為 Domains 的 a tag
print(soup('a', limit=2))                     # 找出前兩個 a tag
```
